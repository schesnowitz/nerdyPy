```python

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-5, 6)
y = x**2

fig, ax = plt.subplots()

ax.plot(x, y, "ko-", markerfacecolor="b", markersize=18)
ax.annotate("This is where the shit \n hit the fan!", (x[3], y[3]), 
            fontsize=10, arrowprops=dict(), xycoords='data', 
            xytext=(0, 15), horizontalalignment='center')
ax.grid()
plt.show()
```

```python
y = x**2 + np.cos(x)*10

minpoint = np.argmin(y)

text = f"{x[minpoint]} y{minpoint}"

fig, ax = plt.subplots()
ax.plot(x, y, 'ko-', markerfacecolor='teal', markersize=15)
ax.annotate(text, (x[minpoint], y[minpoint]), fontsize=15,
            xycoords='data',xytext=(x[int(len(x)/2)], 20), 
            horizontalalignment='center', 
            arrowprops=dict(color='red', arrowstyle='wedge'))
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Title")
plt.show()
```
