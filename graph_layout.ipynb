```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 101)

plt.plot(x, x, label= "y=x")
plt.plot(x, x**2, label= "y=x**2")
plt.plot(x, x**3, label= "y=x**3")

# X axis
plt.xlabel('the x axis label')
plt.xlim([x[0], x[-1]])

# Y axis
plt.ylabel('the y axis label')
plt.ylim([-10, 10])

plt.title("The Title Goes Here!")

# get the graph as a square
print(plt.gca().get_data_ratio())
plt.gca().set_aspect(1 / plt.gca().get_data_ratio())

plt.legend() # remember to activate the legend!
# plt.legend("y=x", "y=x**2" ,"y=x**3")other option to show legend

plt.grid()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(x, x, label= "y=x", color=[10/255, 242/255, 227/255])
ax.plot(x, x**2, label= "y=x**2", color=[238/255, 242/255, 10/255])
ax.plot(x, x**3, label= "y=x**3", color=[10/255, 242/255, 37/255])

ax.legend()

ax.set_xlabel("x label")
ax.set_ylabel("y label")

ax.set_title("The Title")

ax.set_xlim(-3, 3)
ax.set_ylim(-10, 10)
ax.grid()
plt.show()
```
