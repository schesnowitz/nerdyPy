```python
import matplotlib.pyplot as plt
import numpy as np


plt.subplot(1, 2, 1)
plt.plot(np.random.randn(10))

plt.subplot(1, 2, 2)
plt.plot(np.random.randn(10))

plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(np.random.randn(10))

plt.subplot(2, 1, 2)
plt.plot(np.random.randn(10))

plt.show()
```

```python
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.plot(np.random.randn(7))
```

```python
fig,ax = plt.subplots(1, 3)
x = np.arange(10)

ax
# returns array of plots
```

```python
fig,ax = plt.subplots(1, 3, figsize=(15, 2))
x = np.arange(10)

ax[0].plot(x, x**2, "y")
ax[1].plot(x, np.sqrt(x), "r")
ax[2].plot(x, x**2+25, "m")


plt.show()
```

```python
fig, ax = plt.subplots(2, 2)
ax[0, 0].plot(np.random.randn(10, 4))
ax[0, 1].plot(np.random.randn(10, 4))
ax[1, 0].plot(np.random.randn(10, 4))
ax[1, 1].plot(np.random.randn(10, 4))

plt.show()
```
